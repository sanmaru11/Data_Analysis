boolean indexing: 배열의 각 요소의 선택 여부를 Boolean mask를 이용하여 지정하는 방식

```python
x = np.arange(7)

# Boolean mask? True, False로 구성된 mask array

print(x)
print(x < 3)
print(x > 7)

print(x[x<3])
print(x[x % 2 == 0])
```

```python
import numpy as np
import pandas as pd

# 두 개의 시리즈 데이터가 있습니다.
print("Population series data:")
population_dict = {
    'korea': 5180,
    'japan': 12718,
    'china': 141500,
    'usa': 32676
}
population = pd.Series(population_dict)
print(population, "\n")

print("GDP series data:")
gdp_dict = {
    'korea': 169320000,
    'japan': 516700000,
    'china': 1409250000,
    'usa': 2041280000,
}
gdp = pd.Series(gdp_dict)
print(gdp, "\n")


# 이곳에서 2개의 시리즈 값이 들어간 데이터프레임을 생성합니다.
print("Country DataFrame:")
country = pd.DataFrame({
'population' : population,
'gdp':gdp
})

print(country)
```

```python
country.loc['china'] # 인덱스가 china인것의 데이터를 구하라
country.loc['korea' : 'japan', :'population'] # 인덱스는 japan부터 korea까지 :population값 출력
```

```python
country.iloc[0] # 암묵적으로 순서를 통해 데이터 출력
country.iloc[1:3 , :2]
```

```python
country['gdp'] # 컬럼명을 활용하여 데이터 선택 가능 - 시리즈

country[['gdp']] # 데이터프레임 그 자체로 나옴
```

```python
country[country['population'] < 10000] #masking 연산

country.query("population > 10000") # query 함수 활용
```

```python
gdp_per_capita = country['gdp'] / country['population'] # 시리즈도 연산자 활용 가능
country['gdp per capita'] = gdp_per_capita
print(country)
```

```python
import numpy as np
import pandas as pd

print("Masking & query")
df = pd.DataFrame(np.random.rand(5, 2), columns=["A", "B"])
print(df, "\n")

# 데이터 프레임에서 A컬럼값이 0.5보다 작고 B컬럼 값이 0.3보다 큰값들을 구해봅시다.
# 마스킹 연산을 활용하여 출력해보세요!
print(df[(df['A'] < 0.5) & (df['B'] > 0.3)])


# query 함수를 활용하여 출력해보세요!

print(df.query("A < 0.5 and B > 0.3"))
```

```python
import numpy as np
import pandas as pd

data = {
    'korean' : [50, 60, 70],
    'math' : [10, np.nan, 40]
}
df = pd.DataFrame(data, index = ['a','b','c'])
print(df, "\n")

# 각 컬럼별 데이터 개수
col_num = df.count(axis = 0)
print(col_num, "\n")

# 각 행별 데이터 개수
row_num = df.count(axis = 1)
print(row_num, "\n")

# 각 컬럼별 최댓값
col_max = df.max(axis = 0)
print(col_max, "\n")

# 각 컬럼별 최솟값
col_min = df.min(axis = 0)
print(col_min, "\n")

# 각 컬럼별 합계
col_sum = df.sum(axis = 0)
print(col_sum, "\n")

# 컬럼의 최솟값으로 NaN값 대체
math_min = df['math'].min()
df['math'] = df['math'].fillna(math_min)
print(df, "\n")

# 각 컬럼별 평균
col_avg = df.mean(axis = 0)
print(col_avg, "\n")
```

```python
df = pd.DataFrame({
    'data1' : range(6),
    'data2' : [4, 4, 6, 0, 6, 1],
    'key' : ['A', 'B', 'C', 'A', 'B', 'C']
})
df
```

```python
df.groupby('key').sum()
```

```python
df.groupby(['key', 'data1']).sum()
```

aggregate : groupby를 통해서 집계를 한번에 게산하는 방법

```python
df.groupby('key').aggregate(['min', np.median, max])
```

```python
df.groupby('key').aggregate({'data1' : 'min', 'data2' : np.sum})
```

filter : groupby를 통해서 그룹 속성을 기준으로 데이터 필터링

```python
def filter_by_mean(x):
  return x['data2'].mean() > 3 # data2에 대해 평균이 3보다 크면 True

df
```

```python
df.groupby('key').mean()
```

```python
df.groupby('key').filter(filter_by_mean) # filter_by_mean 에서 True가 나온 값들만 반환
```

apply.lambda : groupby를 통해서 묶인 데이터에 함수 적용

```python
df.groupby('key').apply(lambda x : x.max() - x.min())
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'key': ['A', 'B', 'C', 'A', 'B', 'C'],
    'data1': [1, 2, 3, 1, 2, 3],
    'data2': [4, 4, 6, 0, 6, 1]
})
print("DataFrame:")
print(df, "\n")

# groupby 함수를 이용해봅시다.
# key를 기준으로 묶어 합계를 구해 출력해보세요.
sum = df.groupby('key').sum()
print(sum)
# key와 data1을 기준으로 묶어 합계를 구해 출력해보세요.
sum_data1 = df.groupby(['key', 'data1']).sum()
print(sum_data1)
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'key': ['A', 'B', 'C', 'A', 'B', 'C'],
    'data1': [0, 1, 2, 3, 4, 5],
    'data2': [4, 4, 6, 0, 6, 1]
})
print("DataFrame:")
print(df, "\n")

# aggregate를 이용하여 요약 통계량을 산출해봅시다.
# 데이터 프레임을 'key' 칼럼으로 묶고, data1과 data2 각각의 최솟값, 중앙값, 최댓값을 출력하세요.
df_1 = df.groupby('key').aggregate([min, np.median, max])
print(df_1)

# 데이터 프레임을 'key' 칼럼으로 묶고, data1의 최솟값, data2의 합계를 출력하세요.
df_2 = df.groupby('key').aggregate({'data1' : 'min', 'data2' : np.sum})
print(df_2)
```
